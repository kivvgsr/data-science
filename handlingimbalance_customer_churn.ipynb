```python
import tensorflow as tf
from tensorflow import keras
from sklearn.metrics import confusion_matrix , classification_report
```

```python
def ANN(X_train, y_train, X_test, y_test, loss, weights):
    model = keras.Sequential([
        keras.layers.Dense(26, input_dim=26, activation='relu'),
        keras.layers.Dense(15, activation='relu'),
        keras.layers.Dense(1, activation='sigmoid')
    ])

    model.compile(optimizer='adam', loss=loss, metrics=['accuracy'])

    if weights == -1:
        model.fit(X_train, y_train, epochs=100)
    else:
        model.fit(X_train, y_train, epochs=100, class_weight = weights)

    print(model.evaluate(X_test, y_test))

    y_preds = model.predict(X_test)
    y_preds = np.round(y_preds)

    print("Classification Report: \n", classification_report(y_test, y_preds))

    return y_preds
```

```python
y_preds = ANN(X_train, y_train, X_test, y_test, 'binary_crossentropy', -1)
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=15, stratify=y)
```

```python
y_train.value_counts()
```

```python
df3  = X_train.copy()
df3['Churn'] = y_train
```

```python
df3_class0 =  df3[df3.Churn == 0]
df3_class1 =  df3[df3.Churn == 1]
```

```python

df3_class0.shape,df3_class1.shape
```

```python
def get_majority(start,end):
  df_train = pd.concat([df3_class0[start:end],df3_class1], axis = 0)
  x_train = df_train.drop('Churn',axis ='columns')
  y_train =  df_train.Churn
  return x_train , y_train
```

```python
x_train , y_train =  get_majority(0,1495)
y_preds1 = ANN(x_train, y_train, X_test, y_test, 'binary_crossentropy', -1)
```

```python
x_train , y_train =  get_majority(1495,2990)
y_preds2 = ANN(x_train, y_train, X_test, y_test, 'binary_crossentropy', -1)
```

```python
x_train , y_train =  get_majority(2990,4130)
y_preds3 = ANN(x_train, y_train, X_test, y_test, 'binary_crossentropy', -1)
```

# in this method 4 we take the majority of the votes  if ⁉
vote0 =0
vote1 = 1
vote2 = 0
then majority is 0


if ⁉
vote0=0
vote1=1

henc e we conclude that the if sum of the votes is > 1 then the answer is 1 , else 0

```python
y_ped_final = y_preds1.copy()
for i in range(len(y_preds1)):
  total = y_preds1[i]+y_preds2[i]+y_preds3[i]

  if total > 1 :
    y_ped_final[i] = 1
  else:
    y_ped_final[i] =0
```

```python
print(classification_report(y_test,y_ped_final))
```
